6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Shape3D:

    def get_volume(self):
        return 0

    def __add__(self, other):
        return self.get_volume() + other.get_volume()


class Cylinder(Shape3D):

    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def get_volume(self):
        return math.pi * self.radius ** 2 * self.height


class Cube(Shape3D):

    def __init__(self, side):
        self.side = side

    def get_volume(self):
        return self.side ** 3


cylinder = Cylinder(3, 5)
cube = Cube(4)

total_volume = cylinder + cube

print("Об'єм циліндра:", cylinder.get_volume())
print("Об'єм куба:", cube.get_volume())
print("Сумарний об'єм:", total_volume)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:

    def __init__(self, wallet_id, balance=0):
        self.__id = wallet_id
        self.__balance = balance

    @property
    def wallet_id(self):
        return self.__id

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, amount):
        if amount < 0:
            print("Баланс не може бути від'ємним!")
        else:
            self.__balance = amount

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Поповнення на {amount}. Баланс: {self.__balance}")
        else:
            print("Сума поповнення має бути більшою за 0")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Недостатньо коштів!")
        elif amount <= 0:
            print("Сума має бути більшою за 0")
        else:
            self.__balance -= amount
            print(f"Знято {amount}. Баланс: {self.__balance}")


wallet = EWallet("ABC123", 100)

wallet.deposit(50)
wallet.withdraw(30)

print("ID гаманця:", wallet.wallet_id)
print("Баланс:", wallet.balance)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} — {self.price} грн"


class ShoppingCart:

    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"{item.name} додано до кошика")

    def remove_item(self, item_name):
        for item in self.items:
            if item.name == item_name:
                self.items.remove(item)
                print(f"{item_name} видалено з кошика")
                return
        print("Товар не знайдено")

    def show_items(self):
        if not self.items:
            print("Кошик порожній")
        else:
            print("Товари у кошику:")
            for item in self.items:
                print(item)


cart = ShoppingCart()

item1 = Item("Ноутбук", 30000)
item2 = Item("Мишка", 500)
item3 = Item("Клавіатура", 1200)

cart.add_item(item1)
cart.add_item(item2)
cart.add_item(item3)

cart.show_items()

cart.remove_item("Мишка")
cart.show_items()
```

10. (13б). Створіть клас Vehicle з атрибутами brand (manufacturer) та production_age. Створіть ще один клас Engine з такими атрибутами, як type (ex. petrol) та size (volume, litres). Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vehicle:

    def __init__(self, brand, production_age):
        self.brand = brand
        self.production_age = production_age


class Engine:

    def __init__(self, engine_type, size):
        self.engine_type = engine_type
        self.size = size


class Car(Vehicle, Engine):

    def __init__(self, brand, production_age, engine_type, size, model):
        Vehicle.__init__(self, brand, production_age)
        Engine.__init__(self, engine_type, size)

        self.model = model

    def info(self):
        print(f"Car: {self.brand} {self.model}")
        print(f"Age: {self.production_age} years")
        print(f"Engine: {self.engine_type}, {self.size}L")


car = Car("Toyota", 5, "petrol", 2.0, "Camry")

car.info()

print("MRO:", Car.mro())
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:

    def __init__(self):
        self.days = ["Понеділок", "Вівторок", "Середа", "Четвер", "Пʼятниця", "Субота", "Неділя"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration


week = WeekDays()

for day in week:
    print(day)
```
