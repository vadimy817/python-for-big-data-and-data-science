1. Створити змінні, що посилаються на два цілих числа, що однакові за значенням, де значення належить проміжку від -5 до 256. Перевірте, чи будуть ці змінні рівні тільки за значенням або ж ще будуть посилатися на один і той самий обʼєкт в памʼяті? Наведіть код та дайте текстову відповідь нижче.

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
print(my_set)
my_set.add(10)
my_set.add(3.5)
my_set.add("hello")
my_set.add(True)
print(my_set)
```

*місце для текстової відповіді Так множини — змінний (mutable) тип даних.
Можна додавати (add()), видаляти (remove(), discard()), очищати (clear())
Але елементи множини мають бути незмінними (immutable), наприклад int, float, str, tuple.
Не можна додати список чи словник у множину, бо вони змінні.

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_set = set(my_list)
print(my_set)
```

*місце для текстової відповіді Множина (set) в Python не допускає повторів

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
union_set = set1.union(set2)
print("Об'єднання:", union_set)
diff_set = set1.difference(set2)
print("Різниця (set1 - set2):", diff_set)
intersection_set = set1.intersection(set2)
print("Перетин:", intersection_set)
sym_diff_set = set1.symmetric_difference(set2)
print("Симетрична різниця:", sym_diff_set)
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
my_dict = {}
my_dict[1] = "integer key"
my_dict["name"] = "string key"
my_dict[3.14] = "float key"
my_dict[(1, 2)] = "tuple key"
print(my_dict)
```

*місце для текстової відповіді Ні, список не може бути ключем словника. Ключ словника повинен бути хешованим. Список змінний тому його вміст може змінюватися і Python не може гарантувати стабільний хеш.
Тому список як ключ викликає помилку.
